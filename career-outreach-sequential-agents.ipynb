# ðŸŽ¯ Career-Targeted Cold Outreach Sequential Agent Workflow

**Build personalized LinkedIn outreach messages using sequential agents.**

This notebook demonstrates a practical application of sequential agents to generate career-targeted cold outreach messages. The workflow:

1. **Problem Finder Agent** - Identifies key challenges and problems in a specific industry
2. **Problem Solver Agent** - Maps how your career skills can solve those problems
3. **Cold Outreach Message Agent** - Generates personalized LinkedIn messages for each opportunity

This sequential workflow ensures deterministic execution where each agent's output feeds into the next, creating a reliable pipeline for career networking and job searching.

**â„¹ï¸ Use Case:** Career switchers, job seekers, and professionals looking to build targeted networking strategies.

---

## âš™ï¸ Section 1: Import Required Libraries

Import all necessary components from the Agent Development Kit and configure the environment.

```python
import os
from dotenv import load_dotenv
from google.adk.agents import Agent, SequentialAgent
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types

print("âœ… All required libraries imported successfully.")
```

---

## ðŸ”‘ Section 2: Configure Gemini API Authentication

Load your API key from the `.env` file and set up authentication for the Gemini API.

```python
# Load environment variables from .env file
load_dotenv()

try:
    GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
    if not GOOGLE_API_KEY:
        raise ValueError("GOOGLE_API_KEY not found in .env file")
    os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
    os.environ["GOOGLE_GENAI_USE_VERTEXAI"] = "FALSE"
    print("âœ… Gemini API key setup complete. Ready to build agents!")
except Exception as e:
    print(f"ðŸ”‘ Authentication Error: Please make sure you have added 'GOOGLE_API_KEY' to your .env file. Details: {e}")
```

---

## ðŸ” Section 3: Define the Problem Finder Agent

This agent identifies key challenges and problems facing a specific industry. It acts as the first step in our pipeline, discovering opportunities where your skills might be valuable.

**Input:** User's career and the target industry
**Output:** A clear, numbered list of 5-7 current problems/challenges in that industry

```python
problem_finder_agent = Agent(
    name="ProblemFinderAgent",
    model="gemini-2.5-flash-lite",
    instruction="""You are an industry research specialist. Your job is to identify current problems and challenges facing a specific industry.

Based on the user's target industry, research and identify:
1. The 5-7 most pressing problems or challenges currently facing that industry
2. For each problem, provide a brief explanation (1-2 sentences)
3. Format as a clear, numbered list

Be specific and data-driven. Focus on real, current issues that companies in this industry are actively trying to solve.

Your output should ONLY be the numbered list of problems. Do not include any introduction or explanation text.""",
    tools=[google_search],
    output_key="industry_problems",  # This output will be passed to the next agent
)

print("âœ… Problem Finder Agent created successfully!")
```

---

## ðŸ› ï¸ Section 4: Define the Problem Solver Agent

This agent receives the list of industry problems from the Problem Finder and maps them to your specific career skills. It identifies which problems you could potentially help solve.

**Input:** Industry problems (from Problem Finder Agent) + user's career background
**Output:** A structured list of solvable problems with potential solutions based on your skills

```python
problem_solver_agent = Agent(
    name="ProblemSolverAgent",
    model="gemini-2.5-flash-lite",
    instruction="""You are a career strategy specialist. Your job is to analyze industry problems and map them to someone's career skills.

You have been given the following industry problems:
{industry_problems}

Now, for each problem listed above, analyze how the user's career skills and experience could help solve it. 

Your task:
1. Review each industry problem
2. For problems where the user's background is relevant, identify how their skills apply
3. Create a concise "Problem-Solution Match" entry for each solvable problem
4. Format: 
   Problem: [Name of problem]
   How your skills help: [2-3 sentences on how the user's career background addresses this]
   Potential value: [Brief description of the unique value they bring]

Only include problems where there's a clear connection to the user's career background. Skip problems that don't align with their skills.

Output should be a clear list of "Problem-Solution Match" entries only. No introduction or extra text.""",
    output_key="solvable_problems",  # This output will be passed to the next agent
)

print("âœ… Problem Solver Agent created successfully!")
```

---

## ðŸ’¬ Section 5: Define the Cold Outreach Message Agent

This is the final agent in our pipeline. It takes each solvable problem from the Problem Solver Agent and generates personalized, concise LinkedIn connection request messages. Each identified opportunity gets its own tailored message.

**Input:** List of solvable problems with solutions (from Problem Solver Agent)
**Output:** One personalized LinkedIn message for each identified opportunity

```python
cold_outreach_agent = Agent(
    name="ColdOutreachAgent",
    model="gemini-2.5-flash-lite",
    instruction="""You are a professional LinkedIn outreach specialist. Your job is to create personalized, compelling cold outreach messages.

You have been given a list of problems that match the user's career skills:
{solvable_problems}

Your task:
1. For EACH problem-solution match listed, create ONE unique LinkedIn connection request message
2. Each message should:
   - Be 40-60 words (concise, not too long)
   - Reference the specific problem mentioned
   - Subtly highlight how the user's skills address it
   - End with a genuine, non-pushy call to action
   - Sound natural and professional, not generic
   - Include a personalization hook (mention their industry/challenge)

3. Format your output as:
   **Message 1:**
   [LinkedIn message text]
   
   **Message 2:**
   [LinkedIn message text]
   
   And so on for each problem-solution match

Important: Create one message per problem identified. Make each message unique and tailored to that specific problem.
Do not include any introduction or explanation outside the messages themselves.""",
    output_key="linkedin_messages",  # Final output of the entire pipeline
)

print("âœ… Cold Outreach Message Agent created successfully!")
```

---

## ðŸ”— Section 6: Create the Sequential Workflow

Now we combine all three agents into a `SequentialAgent` that runs them in order:

1. **Problem Finder Agent** â†’ Identifies industry problems
2. **Problem Solver Agent** â†’ Maps problems to your skills  
3. **Cold Outreach Agent** â†’ Generates personalized messages

The output of each agent automatically becomes available to the next through the `output_key` mechanism, creating a clean, deterministic pipeline.

```python
# Create the Sequential Agent that chains all three agents together
career_outreach_pipeline = SequentialAgent(
    name="CareerOutreachPipeline",
    sub_agents=[
        problem_finder_agent,      # Step 1: Find industry problems
        problem_solver_agent,      # Step 2: Map problems to skills
        cold_outreach_agent        # Step 3: Generate outreach messages
    ],
)

print("âœ… Sequential Agent Pipeline created!")
print("Pipeline flow: Problem Finder â†’ Problem Solver â†’ Cold Outreach Messages")
```

## ðŸš€ Section 7: Configure Your Career & Industry

Simply set your career background and target industry as variables below. The pipeline will automatically handle the rest!

ðŸ‘‡ **Modify only these two variables:**

```python
# ===== SET YOUR CAREER AND INDUSTRY HERE =====
# Simply fill in these two variables and run the cell

YOUR_CAREER_BACKGROUND = """Senior Full-Stack Software Engineer with 5+ years of experience in building scalable web applications, system design, and team leadership. Skills include Python, JavaScript, React, Django, microservices architecture, and cloud deployment (AWS)."""

TARGET_INDUSTRY = "Marketting"

# ===== NO NEED TO MODIFY ANYTHING BELOW THIS LINE =====

# Initialize the runner
runner = InMemoryRunner(agent=career_outreach_pipeline)

# Automatically construct the prompt from your variables
prompt = f"""
My career background: {YOUR_CAREER_BACKGROUND}

Target industry: {TARGET_INDUSTRY}

Please identify the key problems and challenges in the {TARGET_INDUSTRY} industry where my skills could be valuable, then generate personalized LinkedIn outreach messages for each identified opportunity.
"""

# Run the pipeline
response = await runner.run_debug(prompt)

print(f"\nâœ… Pipeline completed!")
print(f"Career: {YOUR_CAREER_BACKGROUND[:50]}...")
print(f"Industry: {TARGET_INDUSTRY}")
```
